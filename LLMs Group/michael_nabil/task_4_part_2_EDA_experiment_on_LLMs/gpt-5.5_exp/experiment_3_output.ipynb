```
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
df=pd.read_csv('T1.csv')
df.head()
```

```
df.info()
```

```
num=df.select_dtypes('number')
num.hist(figsize=(12,8));plt.show()
```

```
if len(num.columns)>1:
    sns.heatmap(num.corr(),annot=False,cmap='coolwarm');plt.show()
```

```
for c in num.columns[:5]:
    plt.figure();plt.plot(df[c]);plt.title(c);plt.show()
```
